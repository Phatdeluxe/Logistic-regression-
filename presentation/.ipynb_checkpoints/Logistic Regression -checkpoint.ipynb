# Logistic Regression 
### Linear Boundary 

What is the core outcome of the Logistic regression?

    a linear boundary between the classes (in this case default vs. no_default)

The following shows the outcome of Logistic regression for this data

```python
img2 = 'default_data_with_linear_boundary.png'
Image(url=img2)
```

ANY point to the right side of the line will be predicted as default and ANY point to left is predicted to be no_default

The data clearly shows that we are not going to be 100% accurate


### How do I compute the linear boundary?
In live coding, we will use sklearn package together to fit a logistic regression. The complete theory (i.e. Maximum likelihood method with logit assumption) is beyond the scope of this class. Interested students, please read section 4.3.2 of "An Introduction to Statistical Learning with Applications in R" "http://faculty.marshall.usc.edu/gareth-james/ISL/". 


The theory provides useful properties for logistic regression. For example, for each point we can estimate the probability (or likelihood) of being default or no_default. 

### How do I compute probabilities ?

#### Intuition:

    The probabilities are related to the distance from the boundary: 
    
    Farther distance to right: Pr(y=default) increases
    Farther distance to left: Pr(y=no_default) increases
    and always Pr(y=default)+Pr(y=no_default) =1
    

```python
img3 = 'default_data_z_countors.png'
Image(url=img3)
```

#### Formulation

Assume the boundary is denoted by 

$$0 = {const.+\alpha * balance + \beta*income}$$

Then parallel lines can defined as(in some textbooks the following line is called Logit function):

$$z = {const.+\alpha * balance + \beta*income}$$

Where z can be interpreted as a distance from the boundary .

Logistic regression uses logistic function to compute the likelihood:
$$pr(default=yes) = f(z) = \frac{1}{1+exp(-z)}$$

$$ pr(default=no) = 1-f(z)$$

```python
img4 = 'logistic_function.png'
Image(url=img4)
```

Some characteristics of the logistic function: 

1- when z=0, it should return 0.5

2- When we are very far from the boundary, the likelihood should not change alot


```python
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import pandas as pd
```

I use Sklearn package https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html 

```python
# load data
train = pd.read_csv('train.csv', index_col=0) 
# determine Features
X= train[['balance', 'income']]
# detremine lables
y = train['label']
# create classfier object. Parameter values are set at this stage. You do not need to worry about parameters, but
# keep in mind the default parameter values for logistic regression would not give you plain vanila logistic regression
# The following parameter would give you that behavior. 
lr = LogisticRegression(C=1e20, solver='liblinear', penalty='l1')
# fir the model
lr.fit(X, y)
```

```python
# predict classes
y_pred = lr.predict(X)
# predict probabilities 
prob_pred = lr.predict_proba(X)
```

```python
fig, (ax1, ax2, ax3) =plt.subplots(3,1, figsize=(10, 10))

ax1.scatter(X.balance, X.income, c=y)
ax2.scatter(X.balance, X.income, c=y_pred)
ax3.scatter(X.balance, X.income, c=prob_pred[:,1])
ax1.set_title("Colors show observed classes", size=15)
ax2.set_title("Colors show predicted classes", size=15)
ax3.set_title("Colors show predicted probabilities", size=15)
for ax in (ax1, ax2, ax3): 
    ax.set_xlabel('balance')
    ax.set_ylabel('income')

plt.tight_layout()
```

How do I summarize the model performance?
    In this especial example, we use accuracy (fraction of observations that are correctly classified)
    
    Important Note: when the two classes are not balanced in terms of observation or importance, we do NOT use accuracy; We will discuss it in the next classes. 
    

```python
# compute accuracy
lr.score(X, y)
```
